# Document AI Analyzer - Triagem Sintetica

Exemplo simples de priorizacao de documentos para analise assistida por IA.

```python
import pandas as pd
df = pd.read_csv('../data/documentos_exemplo.csv')
peso_complexidade = {'baixa': 1, 'media': 2, 'alta': 3}
df['prioridade_analise'] = df['complexidade'].map(peso_complexidade) * 10 + df['prazo_dias'].apply(lambda x: 30 if x > 0 and x <= 20 else 10 if x > 20 else 0)
df.sort_values('prioridade_analise', ascending=False)[['documento','tipo','tema_principal','prioridade_analise','acao_recomendada']]
```
